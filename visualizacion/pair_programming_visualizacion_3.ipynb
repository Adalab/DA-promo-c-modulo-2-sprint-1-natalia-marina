**EJERCICIO 1** 

Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.

```python
df= pd.read_csv('attacks5.csv')
```

```python
df.head(1)
```

```python
df.info()
```

```python
df_2015 = df[df['year'] >= 2015]
df_2015.head(2)
```

```python
plt.figure(figsize =(14,14))

fig = sns.histplot(y='activity', data = df_2015)

fig.set(ylabel='Actividades', xlabel = 'Número')

fig.set_title('Histograma de las actividades desde 2015' )

plt.show()
```

En esta gráfica se puede observar que desde el año 2015 la actividad más practicada es el surf puede que debido al boom de ese deporte en ese año, 
seguido de la natación, el wading, spearfishing, fishing y snorkeling, el resto de actividades son practicadas por poca cantidad de personas.

**EJERCICIO 2**

Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. Hacedlo solo desde el 2015.

```python
df_2015.head(2)
```

```python
plt.figure(figsize =(10,5))

fig = sns.countplot(x='sex_', data = df_2015, hue = 'year')

fig.set(xlabel='Género', ylabel = 'Ataques')

fig.set_title('Histograma de los ataques por sexo desde 2015' )

plt.legend(bbox_to_anchor=(1,1))

plt.show()
```

Podemos observar que desde el año 2015 el género másculino ha sufrido más número de ataques en comparación con el género femenino, puede que debido a la tendencia social de que los hombres practiquen más deportes que las mujeres. En los siguientes años, el número de ataques ha ido descenciendo, el motivo podría ser la prevención e información del peligro de los ataques de tiburones.

**EJERCICIO 3**

¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.

```python
# Filtramos desde el año 1500 

df_1500 = df[df['year'] >= 1500]
df_1500.head(2)
```

```python
plt.figure(figsize =(10,30))

fig = sns.countplot(y='year', data = df_1500)

fig.set(xlabel= 'Ataques', ylabel='Año')

fig.set_title('Número de ataques por año desde 1500')

plt.show()
```

En este caso podemos observar y sugerir:
- Que con el paso de los años y el aumento de las tecnologias se lleva un mayor registro y control de los ataques.
- El aumento gradual puede ir relacionado con el aumento del ocio y el turismo a las zonas marítimas.

**EJERCICIO 4**

Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
# Filtramos por los paises, expecificando en el método .isin() cuales de ellos.

df_paises =df[df['country'].isin(['USA','AUSTRALIA', 'SOUTH AFRICA'])]
df_paises.head(2)
```

```python
# Volvemos a filtrar con el df anterior los ataques desde el año 1990

df_paises_1900 = df_paises[df_paises['year'] >=1900.0]
df_paises_1900.head(2)
```

```python
plt.figure(figsize =(20,14))

fig = sns.countplot(x='country',hue = 'year', data = df_paises_1900)

fig.set(xlabel='Pais', ylabel = 'Ataques')

fig.set_title('Ataques por paises desde 1900')

fig.legend(bbox_to_anchor =(1,1),ncol = 4)

plt.show()
```

A partir de esta gráfica podemos sacar las siguientes conclusiones:
- El aumento de la temperatura con el paso de los años y el cambio climático, hace que las aguas sean más calidas y atraigan a más tiburones.
- Más ataques de tiburones en el Atlántico que en el Pacífico debido a mayor número de turismo.
- La sequía en Sudáfrica podría ser una variable que afecte a la disminución del os ataques en comparación con las aguas de Australia y USA
