Cread una columna nueva y una funciÃ³n que nos de el mes en el que ocurriÃ³ el ataque. Tened en cuenta que no todas las filas tienen la misma estructura y que puede que no haya la informaciÃ³n de mes. En ese caso devolved un nulo (NaN).

```python
df = pd.read_csv('attacks5.csv')
df.head(5)
```

```python
df.info()
```

```python
df["date"].unique()
```

```python
#patron = '\w+\-'

#patron = '^\w{3}$-'

patron = '\w{3}\-'
```

```python
# iteramos por cada fila de nuestra columna "date" para sacar los meses con el patrÃ³n previamente esstablecido
```

```python
# Creamos una funciÃ³n, metemos el patron dentro de la funciÃ³n (variable local)

def sacar_mes(col1):
    patron = '\w{3}\-'
    try:
        mes = re.findall(patron, col1)[0]
        return mes.replace("-"," ")
    except: 
        return np.nan
    
```

```python
# Hacer un apply con la funciÃ³n y crear la columna para aÃ±adir el dato
df["mes_ataque"]= df["date"].apply(sacar_mes)
df.sample(10)
```

Cread una columna nueva y una funciÃ³n que devuelva Ãºnicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor.

```python
#Hacemos un unique para ver los valores de la columna. No estÃ¡ limpia, deberemos limpiar estos datoss para que nos sean de utilidad 
df["fatal_(y/n)"].unique()
```

```python
# definimos una funcion para separar los valores vÃ¡lidos. Como hemos observado con el unique() tenemos un ' N'. por ello, haremos una condicion aplicando strip para eliminar el espacio y que este valor nos sea vÃ¡lido
def limpiar (col):
   if col == "N":
      return col 
   if col == "Y":
      return col
   if col == " N":
    return col.strip()
   else:
        return np.nan
```

```python
limpiar("fatal_(y/n)")
```

```python
df["fatal"]= df["fatal_(y/n)"].apply(limpiar)
df.head()
```

Seguid la misma lÃ³gica que en el ejercicio anterior pero con la columna sex. Pero en este caso vamos a incluir un retito mÃ¡s: En este caso querremos tener dos valores Ãºnicos F y M. De nuevo en la columna hay otros carÃ¡cteres raros como 'M ', 'lli', 'N', '.'. En esos casos, queremos que se reemplacen por M o F de forma aleatoria (tendrÃ©is que usar la librerÃ­a random). BONUS: En caso de que sea nulo, que la funciÃ³n devuelva nulo (np.nan).

```python
#veamos que valores tenemos en la columna sex_
df["sex_"].unique()
```

```python
#hagamos una funciÃ³n para que 'lli' se convierta en F/M de manera aleatoria, para ello hemos recurrido a google y encontrado un nuevo mÃ©todo de Random, random.choice()
#primero crearemos una lista con nuestras opciones para pasarsela a este mÃ©todo y que el decida cual poner.

def sustituir_aleatorio (col):
    sexos = ["F","M"]
    if col == "F":
        return col
    elif col == "M":
        return col
    elif col == "lli":
       col = (np.random.choice(sexos))
       return col
       
```

```python
sustituir_aleatorio(["sex_"])
```

```python
df["sexo"]= df["sex_"].apply(sustituir_aleatorio)
df["sexo"].unique()
```
