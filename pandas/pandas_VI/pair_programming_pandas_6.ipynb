**EJERCICIO 1**

¬øCu√°ntos ataques hubo por pa√≠s? Devuelve un DataFrame.
üìå NOTA Ordena los resultados de mayor a menor. Esto todav√≠a no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values(). Aqu√≠ ten√©is una explicaci√≥n de c√≥mo hacerlo.

```python
# Agrupamos los ataques filtrando por la columna country y case_number, utilizando .count para 
# obtener la cantidad

df_paises = df.groupby('country')['case_number'].count()
```

```python
# Utilizamos los paises como √≠ndice con el m√©todo .reset_index

df_paises.reset_index()
```

```python
# Ordenamos de mayor a menor con el m√©todo sort_values

df_paises.sort_values(axis=0, ascending=False, inplace=False,)
```

```python
df_pais = df_paises.reset_index()
df_pais.head()
```

**EJERCICIO 2**

¬øCu√°ntos ataques ha habido por a√±o? Devuelve un DataFrame. De nuevo ordena los resultados de mayor a menor.

```python
df_a√±os = df.groupby('year')['case_number'].agg(ataques=('count'))
df_a√±os.reset_index()
df_a√±os
```

```python
df_a√±os.sort_values(by='ataques', ascending=False, inplace=False,)
df_a√±os.reset_index()
```

```python
df_anyo= df_a√±os.reset_index()
```

```python
df_anyo.head()
```

**EJERCICIO 3**

¬øCu√°ntos ataques hubo por a√±o y sexo? Devolved un DataFrame solo del a√±o 2000 en adelante.

```python
df.columns
```

```python
# Filtramos con la suma de los ataques por a√±o y sexo

df_3 = df.groupby(['year', 'sex_'])['case_number'].agg(ataques=('count'))
df_3

```

```python
df_3.reset_index()
```

```python
# Volvemos a filtar por los que han sido a partir del a√±o 2000

df_2000 = df_3[df_3['year'] >= 2000.0]
df_2000
```

**EJERCICIO 4**

Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. ¬øCu√°ntos de los ataques fueron letales por sexo? Devuelve un DataFrame.

```python
# Agrupamos los ataques filtrando por los paises, utilizando el m√©todo isin para USA, AUSTRALIA Y SUDAFRICA 

df_4 = df[df['country'].isin(['USA','AUSTRALIA', 'SOUTH AFRICA'])]
df_4.head(2)
```

```python
# Volvemos a filtrar el por los accidentes que han sido fatales(fatal=Y)

df_fatal = df_4[df_4['fatal_(y/n)']=='Y']
df_fatal.head(2)
```
