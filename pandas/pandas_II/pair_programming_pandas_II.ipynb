```python
import pandas as pd
import numpy as np
import pickle 
```

```python
# Abrimos el archivo csv teniendo en cuenta el encoding y el separador

df_csv = pd.read_csv("Archivos/quejas.csv", encoding= 'latin-1', sep= ';' )
df_csv.head(2)
```

```python
# Abrimos el archivo excel

df_xlsx = pd.read_excel("Archivos/quejas2.xlsx")
df_xlsx.head()
```

```python
# Abrimos el archivo json 

df_json = pd.read_json("Archivos/quejas.json")
df_json.head()
```

```python
# Abrimos el archivo pickle  

df_pickle = pd.read_pickle("Archivos/algas.pickle")
df_pickle.head(5)
```

```python
# Abrimos un clipboard, corriendo este código mientras que tenemos los datos de la tabla excel 
# copiados en el 'portapapeles'

df_clipboard = pd.read_clipboard(sep='\\s+')
df_clipboard.head(2)
```

---

```python
# Guardamos el archivo csv

df_csv.to_csv("Guardado/quejas.csv")
```

```python
# Guardamos el archivo xlsx

df_xlsx.to_excel("Guardado/quejas2.xlsx")
```

```python
# Guardamos el archivo json

df_json.to_json("Guardado/quejas.json")
```

```python
# Guardamos el archivo pickle

df_pickle.to_pickle("Guardado/algas.pickle")
```
